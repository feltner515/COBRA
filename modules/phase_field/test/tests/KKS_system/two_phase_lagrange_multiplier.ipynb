```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({'font.size': 18})
```

```python
Ft = pd.read_csv('two_phase_out_c_0036.csv')
Fm = pd.read_csv('lagrange_multiplier_out_c_0036.csv')
Fa = pd.read_csv('auxkernel_out_c_0040.csv')
Fn = pd.read_csv('nonlinear_out_c_0040.csv')
```

```python
fig = plt.figure(figsize=(15,12))
ax = fig.gca()
ax.plot(Ft['x'], Ft['c'], 'b-', label="two phase", linewidth=5, alpha= 0.25)
ax.plot(Fm['x'], Fm['c'], 'r--', label="multi phase")
ax.plot(Fa['x'], Fa['c'], 'go', label="auxkernel")
ax.plot(Fn['x'], Fn['c'], 'y.', label="nonlinear")
ax.set_xlabel('position')
ax.set_ylabel('concentration')
ax.legend(loc='lower right')
plt.show()
```

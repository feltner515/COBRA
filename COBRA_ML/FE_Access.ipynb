```python
def convert (s):
    new = ""
    for x in s:
        new += x
    return new

def get_variable(dataset):
    nodal_names = []
    for i in range(dataset.variables['name_nod_var'].shape[0]):
        name = convert(str(str(str(nc.variables['name_nod_var'][i].data).split("'")[1:len(nc.variables['name_nod_var'][i].data):2]).split("''")[0]))
        nodal_names.append(name[1:len(name)-2])
    return nodal_names
```

```python
from scipy.interpolate import RegularGridInterpolator
connect = nc.variables['connect1']



X, indices = np.unique(np.array(nc.variables['coordx'][0:len(connect)-1]), return_inverse=True)
Y, indices = np.unique(np.array(nc.variables['coordy'][0:int(len(X)**3)]), return_inverse=True)
Z, indices = np.unique(np.array(nc.variables['coordz'][0:int(len(X)**3)]), return_inverse=True)

x_unsort = np.array(nc.variables['coordx'][0:int(len(X)**3)])
y_unsort = np.array(nc.variables['coordy'][0:int(len(X)**3)])
z_unsort = np.array(nc.variables['coordz'][0:int(len(X)**3)])

xg, yg, zg = np.meshgrid(X, Y, Z)

vals = np.array(nc.variables['vals_nod_var44'])[-1,:]
##search algorithm
searchpoint = np.zeros([3])
vals_idx = np.zeros([xg.shape[0], xg.shape[1], xg.shape[2]])
for n in range (0,len(x_unsort)):
    searchpoint = [x_unsort[n], y_unsort[n], z_unsort[n]]
    x_idx = np.where(xg == searchpoint[0], 1,0)
    y_idx = np.where(yg == searchpoint[1],1, 0)
    z_idx = np.where(zg == searchpoint[2],1 ,0)
    idx = np.where((x_idx == 1) & (y_idx == 1) & (z_idx == 1))
    
    vals_idx[int(idx[0]), int(idx[1]), int(idx[2])] = vals[n]

x_vals = np.linspace(0.375, 0.625, 30)
y_vals = np.linspace(0.375, 0.625, 30)
z_vals = np.linspace(0.75, 1, 30)

interp_values = RegularGridInterpolator(points = (X, Y, Z), values= vals_idx, method='cubic', bounds_error=False, fill_value=0)

```

```python
RVE_x, RVE_y, RVE_z = np.meshgrid(x_vals, y_vals, z_vals, indexing='ij')
RVE_disp = np.zeros(RVE_x.shape)
for i in range(RVE_x.shape[0]):
    for j in range(RVE_y.shape[1]):
        for k in range(RVE_z.shape[2]):
            RVE_disp[i,j,k] = interp_values([RVE_x[i,j,k], RVE_y[i,j,k], RVE_z[i,j,k]])
```
